```python
import pandas as pd
import numpy as np
```

```python
df_Tue = pd.read_csv("Tuesday-WorkingHours.pcap_ISCX.csv")
df_Wed = pd.read_csv("Wednesday-workingHours.pcap_ISCX.csv")
df_ThuMor = pd.read_csv("Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv", encoding='cp1252')
df_FriAft = pd.read_csv("Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv")
```

```python
dataset = pd.concat([df_Tue, df_Wed, df_ThuMor, df_FriAft], axis=0, ignore_index=True)
```

concatenating all the required datasets

axis = 0 refers to change in rows i.e. in this case row-wise concatenation

If ignore_index = True, the source objects are ignored and 0,1,2...n indexes are used in the output

```python
dataset[" Label"].unique()
```

```python
dataset.drop(dataset.loc[dataset[" Label"] == " Label"].index, inplace=True)
```

Remove erroneous ' Label' row

drop() => a function used to remove rows or columns of a dataset

Dataframe.loc => Access a group of rows and columns by label(s) or a boolean array.

When inplace = True, the data is modified in place, which means it will return nothing and the dataframe is now updated. When inplace = False, which is the default, then the operation is performed and it returns a copy of the object. You then need to save it to something.

```python
dataset.replace(to_replace=['FTP-Patator', 'SSH-Patator', 'DoS slowloris',
       'DoS Slowhttptest', 'DoS Hulk', 'DoS GoldenEye', 'Heartbleed',
       'Web Attack – Brute Force', 'Web Attack – XSS',
       'Web Attack – Sql Injection', 'DDoS'], value = "Attack", inplace = True)

dataset.replace(to_replace=['BENIGN'], value = "Benign", inplace = True)
```

Replace the differents types of attacks with a single representation "Attack"

Assuming that "Protocol" is a categorical attribute.

```python
#explicitly converting protocols into strings
dataset = dataset.astype({" Protocol": str})
```

```python
dataset[" Protocol"].unique()
```

```python
dataset = pd.get_dummies(dataset, columns=[' Protocol'], drop_first = True)
```

Performing one hot encoding using pandas i.e. pd.get_dummies()

Then, removing the first column using drop_first=True 
Since one of the columns can be generated completely from the others, and hence retaining this extra column does not add any new information for the modelling process. It reduces the corelation between column variables.

```python
dataset
```

```python
#Keeping label as last column
dataset.insert(len(dataset.columns)-1, " Label", dataset.pop(" Label"))
```

```python
dataset
```

```python
#These features have value 0 througout the table hence, they have very less variance and offer little predictive power.
#They will also generate bad values during feature selection so, they are removed

dataset['Fwd Avg Bytes/Bulk'].unique()
dataset[' Fwd Avg Packets/Bulk'].unique()
dataset[' Fwd Avg Bulk Rate'].unique()
dataset[' Bwd Avg Bytes/Bulk'].unique()
dataset[' Bwd Avg Packets/Bulk'].unique()
dataset['Bwd Avg Bulk Rate'].unique()
dataset[' CWE Flag Count'].unique()
```

```python
#Dropping some unnecessary columns after analysis

columns_to_drop = ['Flow ID', ' Source IP', ' Destination IP', ' Timestamp', ' Destination Port', 'Fwd PSH Flags',
       ' Bwd PSH Flags', ' Fwd URG Flags', ' Bwd URG Flags', 'Flow Bytes/s', ' Flow Packets/s', 'Fwd Avg Bytes/Bulk',
       ' Fwd Avg Packets/Bulk', ' Fwd Avg Bulk Rate', ' Bwd Avg Bytes/Bulk',
       ' Bwd Avg Packets/Bulk', 'Bwd Avg Bulk Rate', ' CWE Flag Count']
```

```python
dataset.drop(columns = columns_to_drop, inplace = True)
```

```python
#Dropping missing or duplicate values
dataset.dropna(inplace = True)
dataset.drop_duplicates(inplace = True)
```

```python
dataset[" Label"].value_counts()
```

```python
#Replacing Benign label as 0 and Attack label as 1

dataset.replace(to_replace = "Benign", value = 0, inplace = True)
dataset.replace(to_replace = "Attack", value = 1, inplace = True)
```

```python
dataset
```

```python
dataset.info()
```

```python
#Undersampling to create a balanced dataset
from sklearn.utils import resample

data_majority = dataset[(dataset[' Label']==0)]
data_minority = dataset[(dataset[' Label']==1)]

data_resample_majority = resample(data_majority,
                           replace=True,
                           n_samples = 5000,
                           random_state=42)

data_resample_minority = resample(data_minority,
                           replace=True,
                           n_samples = 5000,
                           random_state=42)
```

```python
Final_dataset = pd.concat([data_resample_majority, data_resample_minority])
```

```python
Final_dataset[" Label"].value_counts()
```

```python
Final_dataset = Final_dataset.sample(frac=1, random_state=1).reset_index()
```

```python
#Saving the preprocessed data
Final_dataset.to_csv("Test_data_10k.csv", index = False)
```
